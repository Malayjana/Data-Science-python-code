#### IsolationForest

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
data =pd.read_csv("C:\\Users\\Neha Excelr\\DataSets\\iris.csv")
data_encoded=pd.get_dummies(data)
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
```

```python
# predictions
y_pred_outliers = clf.predict(data_encoded)
```

```python
#-1 for outliers and 1 for inliers.
y_pred_outliers
```

```python
data_encoded['Outliers']=y_pred_outliers
```

```python
data_encoded
```

```python
data_encoded[data_encoded.Outliers==-1]
```

```python
data_encoded.drop('Outliers', axis=1,inplace=True)
data_encoded
```

```python
## Let us add a new data point which is outlier
data_encoded.loc[150]=[20,40,30,50,1,0,0]
```

```python
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
# predictions
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
```

```python
data_encoded.shape
```

```python
len(y_pred_outliers)
```

```python
data_encoded['scores']=y_pred_outliers
data_encoded
```

```python
data_encoded[data_encoded.scores==-1]
```

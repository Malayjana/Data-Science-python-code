```python
#Filtering for nouns and verbs only
nouns_verbs = [token.text for token in doc_block if token.pos_ in ('NOUN', 'VERB')]
print(nouns_verbs)
```

```python
#Counting tokens again
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

X = cv.fit_transform(nouns_verbs)
sum_words = X.sum(axis=0)
words_freq=[(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key = lambda x: x[1], reverse=True)
wf_df = pd.DataFrame(words_freq)
wf_df.columns = ['word', 'count']

wf_df[0:10]
```

```python
import pandas
book=pd.read_csv("C:\\Users\\Neha Excelr\\DataSets\\apple.txt",
                 error_bad_lines=False)
book
```

```python
book = [x.strip() for x in book.x] # remove both the leading and the trailing characters
book = [x for x in book if x] # removes empty strings, because they are considered in Python as False
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk import tokenize
sentences = tokenize.sent_tokenize(" ".join(book))
sentences[0:10]
```

```python
sent_df = pd.DataFrame(sentences, columns=['sentence'])
sent_df
```

```python
affinity_scores = afinn.set_index('word')['value'].to_dict()
```

```python
affinity_scores
```

```python
#Custom function :score each word in a sentence in lemmatised form, 
#but calculate the score for the whole original sentence.
nlp = spacy.load('en_core_web_md')
sentiment_lexicon = affinity_scores

def calculate_sentiment(text: str = None):
    sent_score = 0
    if text:
        sentence = nlp(text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_, 0)
    return sent_score
```

```python
# test that it works
#calculate_sentiment(text = 'The class is awesome. But the topic is very sluggish.')
calculate_sentiment(text = 'I love coding, but it is very difficult sometimes, though it is interesting too')
```

```python
sent_df['sentiment_value'] = sent_df['sentence'].apply(calculate_sentiment)
```
